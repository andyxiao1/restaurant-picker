```python
priceRange = []
credit_cards = []
takeout = []
outdoor = []
for row in modDF.iterrows():
    s = row[1]['attributes']
    if s != None:
        priceRange.append(s.get('RestaurantsPriceRange2'))
        credit_cards.append(s.get("BusinessAcceptsCreditCards"))
        takeout.append(s.get("RestaurantsTakeOut"))
        outdoor.append(s.get("OutdoorSeating"))
    else:
        priceRange.append(None)
        credit_cards.append(None)
        takeout.append(None)
        outdoor.append(None)
        
```

```python
outdoorMod = []
for val in outdoor:
    if val != None and val.lower() == 'true':
        outdoorMod.append(1)
    else:
        outdoorMod.append(-1)
        
```

```python
takeoutMod = []
for val in takeout:
    if val != None and val.lower() == 'true':
        takeoutMod.append(1)
    else:
        takeoutMod.append(-1)
        
```

```python
creditMod = []
for val in credit_cards:
    if val != None and val.lower() == 'true':
        creditMod.append(1)
    else:
        creditMod.append(-1)
        
```

```python
priceRangeMod = []
for val in priceRange:
    if val != None and val != 'None':
        print(val)
        priceRangeMod.append(int(val)-2)
    else:
        priceRangeMod.append(0)
```
