```python
import pandas as pd
```

```python
inspection_data = pd.read_csv('Restaurant_Inspections_Open_Data.csv')
```

```python
inspection_data.head(2)
```

```python
inspection_data.columns
```

```python
inspection_data_keep = inspection_data.drop(columns=['Permit_Number', 'State', 'Zip', 'Inspection_Time', 'Employee_ID', 'Permit_Status', 'Record_Updated'])
```

```python
inspection_data_keep['Latitude'] = 0.0
inspection_data_keep['Longitude'] = 0.0
inspection_data_keep.dtypes




    
```

```python
lat_list = []
long_list = []
for row in inspection_data_keep.iterrows():
    location = row[1]['Location_1']
    lat_long = location.split(',')
    rownum = row[0]
    print(type(rownum))
    lat_list.append(round(float(lat_long[0].split('(')[1]),5))
    long_list.append(round(float(lat_long[1].split(')')[0]),5))
    
```

```python
inspection_data_keep['latitude'] = lat_list
inspection_data_keep['longitude'] = long_list
```
